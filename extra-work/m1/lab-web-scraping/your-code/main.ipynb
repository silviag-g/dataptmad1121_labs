#### Display all the image links from Walt Disney wikipedia page.

````python
# This is the url you will scrape in this exercise
url_disney = 'https://en.wikipedia.org/wiki/Walt_Disney'
````

````python
# your code here
html_disney = requests.get(url_disney).content
````

````python
parsed_html_disney = BeautifulSoup(html_disney, "html.parser") 
````

````python
disney_images = parsed_html_disney.find_all('a',{ "class": "image"})
disney_images
````

````python
#'href' es la localizacion de donde se encuentra la imagen, el hipervinculo del un documentos.
for i in disney_images:
    print(i['href']+ '\n')
````

#### Retrieve an arbitary Wikipedia page of "Python" and create a list of links on that page.

````python
# This is the url you will scrape in this exercise
url_python ='https://en.wikipedia.org/wiki/Python' 
````

````python
# your code here
html_python = requests.get(url_python).content
````

````python
parsed_html_python = BeautifulSoup(html_python, "html.parser") 
````

````python
#importamos re de regex, porque para usar re.compile que encuentra todos los href dentro de wikipedia.
python = parsed_html_python.find_all('a',{'href':re.compile('/wiki/Python')})
for p in python:
    print(p['href']+ '\n')
````
